## 5. Использование пайплайнов

```python
target = 'Heart Attack Risk (Binary)'

# Списки для действий
drop_cols = ['Unnamed: 0', 'id', 'Income', 'Obesity', 'Previous Heart Problems']  # Бесполезные + корр + утечка
skewed_cols = ['Cholesterol', 'Triglycerides', 'BMI']  # Для трансформации (log или power)
bin_cols = ['Diabetes', 'Family History', 'Smoking', 'Alcohol Consumption', 'Medication Use']  # Бинарные (impute most_frequent)
cat_ordinal = ['Diet', 'Stress Level', 'Physical Activity Days Per Week']  # Ordinal (encoder)
cat_nominal = ['Gender']  # Nominal (OHE)
num_cols = ['Age', 'Heart rate', 'Exercise Hours Per Week', 'Sedentary Hours Per Day', 'Sleep Hours Per Day', 'Blood sugar', 'CK-MB', 'Troponin']  # Оставшиеся num
```

```python
train_df = train_df.drop(columns=drop_cols, errors='ignore')
```

```python
# Новая фича: Mean blood pressure
train_df['mean_blood_pressure'] = (train_df['Systolic blood pressure'] + train_df['Diastolic blood pressure']) / 2
num_cols.append('mean_blood_pressure')

test_df['mean_blood_pressure'] = (test_df['Systolic blood pressure'] + test_df['Diastolic blood pressure']) / 2
```

```python
# Сплит на очищенных данных
RANDOM_STATE = 42
TEST_SIZE = 0.25
X = train_df.drop(columns=[target])
y = train_df[target]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)
```

```python
# Custom log transform (для skewed >0, +1 для избежания log(0))
def log_transform(X):
    return np.log1p(X)
```

```python
log_transformer = FunctionTransformer(log_transform)
```

```python
nullable_cols = bin_cols + cat_ordinal + [target]
for col in nullable_cols + num_cols:
    dtype_str = str(train_df[col].dtype)
    if dtype_str.startswith('Int') or dtype_str.startswith('Float'):
        train_df[col] = train_df[col].astype('float64')
        print(f'Конвертировал {col} в float64')
```

```python
np.unique(y_train)
```

```python
# Custom to_str для фикса mixed types
def to_str(X):
    return X.astype(str)
to_str_transformer = FunctionTransformer(to_str)
```

```python
# Пайплайн для бинарных
bin_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent', missing_values=np.nan))
])
```

```python
# Пайплайн для числовых (median для skewed)
num_pipeline_standard = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median', missing_values=np.nan)),
    ('scaler', StandardScaler())
])
num_pipeline_minmax = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median', missing_values=np.nan)),
    ('scaler', MinMaxScaler())
])
# Пайплайн для числовых (scaler as placeholder)
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median', missing_values=np.nan)),
    ('scaler', StandardScaler())  # Тюнинг через param_grid
])
```

```python
# Пайплайн для номинальных категориальных
cat_pipeline_onehot = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing', missing_values=np.nan)),
    ('to_str', FunctionTransformer(lambda X: X.astype(str))),
    ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

# Пайплайн для ординальных
cat_pipeline_ordinal = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent', missing_values=np.nan)),
    ('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

```python
# Пайплайн для skewed (impute + log + scale)
skewed_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median', missing_values=np.nan)),
    ('log', log_transformer),
    ('scaler', StandardScaler())  # Тюнинг аналогично
])
```

```python
# Общий препроцессор
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_cols),
        ('skewed', skewed_pipeline, skewed_cols),
        ('bin', bin_pipeline, bin_cols),
        ('cat_onehot', cat_pipeline_onehot, cat_nominal),
        ('cat_ordinal', cat_pipeline_ordinal, cat_ordinal)
    ],
    remainder='drop'
)
```

```python
# Обнови pipe на ImbPipeline
pipe = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42, sampling_strategy='auto')),  # Oversample minority
    ('model', DummyRegressor())
])
```

```python
# Параметры для подбора (уменьшены для скорости, добавлены class_weights)
param_grid = [
    {
        'preprocessor__num': [num_pipeline_standard, num_pipeline_minmax],
        'model': [KNeighborsClassifier()],
        'model__n_neighbors': [3, 5, 7]
    },
    {
        'preprocessor__num': [num_pipeline_standard, num_pipeline_minmax],
        'model': [DecisionTreeClassifier(random_state=RANDOM_STATE, class_weight='balanced')],
        'model__max_depth': [3, 5, 7]
    },
    {
        'preprocessor__num': [num_pipeline_standard, num_pipeline_minmax],
        'model': [LogisticRegression(max_iter=1000, class_weight='balanced')],
        'model__C': [0.1, 1, 10]
    },
    {
        'preprocessor__num': [num_pipeline_standard, num_pipeline_minmax],
        'model': [SVC(probability=True, class_weight='balanced')],
        'model__C': [1, 10],
        'model__kernel': ['linear', 'rbf']
    },
    {
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler()],
        'preprocessor__skewed__scaler': [StandardScaler(), MinMaxScaler()],
        'model': [CatBoostClassifier(random_state=RANDOM_STATE, verbose=0, auto_class_weights='Balanced')],
        'model__iterations': [100, 200],
        'model__depth': [4, 6],
        'model__learning_rate': [0.01, 0.1]
    }
]
```

```python
# Метрика для оценки качества модели
roc_auc_scorer = make_scorer(roc_auc_score, needs_proba=True)
```

```python
grid = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=3,
    error_score='raise'  # Для дебагинга
)
```

```python
%time 
grid.fit(X_train, y_train)
print("Лучшие параметры:", grid.best_params_)
print("roc_auc на валидации:", grid.best_score_)
```

```python
cv_results_df = pd.DataFrame(grid.cv_results_)
top_10 = cv_results_df.sort_values(by='mean_test_score', ascending=False).head(10)

top_10
```
