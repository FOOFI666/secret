```python
def change_type_df(df, list_change):
    for i in list_change:
        # Конвертируем в numeric (строки/нечисла -> NaN)
        df[i] = pd.to_numeric(df[i], errors='coerce')
        df[i] = df[i].astype('Int32')
        print(f'Тип признака {i} был изменен на {df[i].dtype}')
    return df.head(15)
```

```python
list_change = ['Heart Attack Risk (Binary)', 
               'Diabetes', 
               'Family History', 
               'Smoking', 
               'Obesity', 
               'Alcohol Consumption', 
               'Previous Heart Problems', 
               'Medication Use', 
               'Stress Level', 
               'Physical Activity Days Per Week',]
```

```python
change_type_df(train_df, list_change)
```

## 3. Исследование данных

### 3.1 Создание списков для разных величин

```python
# Список для создания графиков countplot (категориальные и дискретные переменные с низким числом уникальных значений)
bar_cols = [
    'Gender',
    'Diabetes',
    'Family History',
    'Smoking',
    'Obesity',
    'Alcohol Consumption',
    'Diet',
    'Previous Heart Problems',
    'Medication Use',
    'Stress Level',
    'Physical Activity Days Per Week',
    'Heart Attack Risk (Binary)'
]

# Список для создания коробчатой диаграммы и графиком распределения данных (непрерывные переменные)
continuous_cols = [
    'Age',
    'Cholesterol',
    'Heart rate',
    'Exercise Hours Per Week',
    'Sedentary Hours Per Day',
    'Income',
    'BMI',
    'Triglycerides',
    'Sleep Hours Per Day',
    'Blood sugar',
    'CK-MB',
    'Troponin',
    'Systolic blood pressure',
    'Diastolic blood pressure'
]

# Список для создание круговой диаграммы (переменные с пропорциональными категориями, подходящими для показа долей, напр. бинарные)
equal_cols = [
    'Gender',
    'Heart Attack Risk (Binary)'
]
```

<b>3.2 Установка настроек для красивых графиков</b>

```python
sns.set_theme(style="whitegrid", palette="husl", font_scale=1.1)
plt.rcParams['figure.figsize'] = (12, 8)  # Размер фигур по умолчанию
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
fig, axes = plt.subplots(4, 3, figsize=(18, 16))
axes = axes.ravel()  # Разворачиваем в 1D для удобства

for i, col in enumerate(bar_cols):
    if i < len(axes):  # На случай, если меньше подграфиков
        sns.countplot(data=train_df, x=col, hue=col, ax=axes[i], palette='husl', legend=False)
        axes[i].set_title(f'Распределение: {col}', fontweight='bold')
        axes[i].tick_params(axis='x', rotation=45 if len(train_df[col].unique()) > 5 else 0)
        for container in axes[i].containers:
            axes[i].bar_label(container, fmt='%.0f')  # Добавляем значения на бары

plt.tight_layout()
plt.suptitle('Countplots для категориальных переменных', y=1.02, fontsize=18, fontweight='bold')
plt.show()
```

```python
fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for i, col in enumerate(continuous_cols):
    if i < len(axes):
        # Boxplot без palette (чтобы избежать предупреждения; используем color для единого цвета)
        sns.boxplot(data=train_df, y=col, ax=axes[i], color='lightblue')
        axes[i].set_title(f'Boxplot: {col}', fontweight='bold')
        axes[i].tick_params(axis='y', rotation=0)

plt.tight_layout()
plt.suptitle('Boxplots для непрерывных переменных', y=1.02, fontsize=18, fontweight='bold')
plt.show()
```

```python
fig, axes = plt.subplots(4, 4, figsize=(20, 16))
axes = axes.ravel()

for i, col in enumerate(continuous_cols):
    if i < len(axes):
        # Histplot без palette (чтобы избежать предупреждения; цвет по умолчанию или через color)
        sns.histplot(data=train_df, x=col, kde=True, ax=axes[i], bins=30, alpha=0.7, color='skyblue')
        axes[i].set_title(f'Распределение: {col}', fontweight='bold')
        axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.suptitle('Гистограммы распределений для непрерывных переменных', y=1.02, fontsize=18, fontweight='bold')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for i, col in enumerate(equal_cols):
    if i < len(axes):
        value_counts = train_df[col].value_counts()
        axes[i].pie(value_counts.values, labels=value_counts.index.astype(str), autopct='%1.1f%%',
                    colors=sns.color_palette('husl', len(value_counts)), startangle=90, textprops={'fontsize': 12})
        axes[i].set_title(f'Доли: {col}', fontweight='bold', fontsize=14)

plt.suptitle('Круговые диаграммы для пропорциональных категорий', y=1.02, fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```
